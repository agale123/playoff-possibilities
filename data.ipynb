```python
import nwslpy
import pandas as pd
```

```python
teams = nwslpy.load_teams()
teams
```

```python
df = nwslpy.load_matches()
df = df[df["season"] == "2023"]
df = df.sort_values("kickoff")
df["kickoff"] = pd.to_datetime(df["kickoff"]).dt.date
df = df[
    ["home_team_id", "away_team_id", "kickoff", "home_team_score", "away_team_score"]
]
df = (
    df.set_index("home_team_id")
    .join(teams[["team_abbreviation"]])
    .rename(columns={"team_abbreviation": "home_team"})
)
df = (
    df.set_index("away_team_id")
    .join(teams[["team_abbreviation"]])
    .rename(columns={"team_abbreviation": "away_team"})
)
df = (
    df[["kickoff", "home_team", "away_team", "home_team_score", "away_team_score"]]
    .sort_values("kickoff")
    .reset_index(drop=True)
)
df
```

```python
df.to_csv("web/data.csv")
```
